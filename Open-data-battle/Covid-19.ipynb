Удаление пустых столбцов и индексов

```python
df=df.drop(columns=['Unnamed: 0','Unnamed: 40','Unnamed: 41'])
```

Удаление строк "For Note"

```python
df['date']=pd.to_datetime(df['date'], errors='coerce')
df=df[df['date'].notnull()]
```

```python
Группировка по месяцам, континентам
```

```python
grupped=df.groupby(['month','year','continent'])['total_cases'].count()
```

```python
df_grupped=pd.DataFrame(grupped).reset_index()
df_grupped
```

Поиск наименьшей корреляции

```python
corr=df_grupped.pivot_table(columns='continent', index=['month','year']).corr().reset_index()
corr['total_cases'].mean().sort_values()
```
