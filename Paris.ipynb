```python
df['numberOfRooms']=df['numberOfRooms'].str.replace('-',' ').apply(lambda x: w2n.word_to_num(x))
```

```python
df['pool']=df['PoolAndYard'].apply(lambda x: int(x.split(' ')[0]=='has'))
df['yard']=df['PoolAndYard'].apply(lambda x: int(x.split(' ')[3]=='has'))
```

```python
df['category'].unique()
df= pd.get_dummies(df, drop_first=True, columns=['category'])
```

```python
df=df.drop(columns=['Unnamed: 0','PoolAndYard'])
```

```python
for col in ['isNewBuilt','hasStormProtector','hasStorageRoom']:
    df[col]=df[col].apply(lambda x: int(x))
```

```python
df.head()
```

```python
clusters_range = range(20,200, 20)
```

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}

for n_clusters in clusters_range:
    model = KMeans(n_clusters=n_clusters).fit(df)
    distortions.append(sum(np.min(cdist(df, model.cluster_centers_,

                      'euclidean'),axis=1)) /df.shape[0])

    inertias.append(model.inertia_)

  

    mapping1[n_clusters] = sum(np.min(cdist(df, model.cluster_centers_,

                 'euclidean'),axis=1)) / df.shape[0]

    mapping2[n_clusters] = model.inertia_
```

```python
plt.figure(figsize=(10,5))
plt.plot(clusters_range, distortions, 'bx-')
plt.xlabel('Values of clusters_range')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(clusters_range, inertias, 'bx-')
plt.xlabel('Values of clusters_range')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```
